# networkx 使用方法

## 创建图

```python
import networkx as nx
G = nx.Graph()
```

## 添加节点

```python
G.add_node(1)

G.add_nodes_from([2, 3])
```

## 添加边

```python
G.add_edge(2, 3)

G.add_edges_from([(1, 2), (1, 3)])
```

## 绘制图

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=True, font_weight='bold')
```
